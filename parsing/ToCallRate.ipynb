```python
import pandas as pd
from scipy import stats
from statsmodels.formula.api import ols
```

```python
raw = pd.read_csv('nature-alddist-counts.csv', index_col=0, header=0)
raw = raw.drop(['nan'], axis=1)
raw.head()
```

```python
calls = []
callRates = []
natures = []
locs = []

df = raw.copy()
for i, row in df.iterrows():
    if i != 'All':
        for col in df.columns:
            calls.append(df.loc[i, col])
            natures.append(i)
            locs.append(col)
            df.loc[i, col] /= df.loc['All', col]
            callRates.append(df.loc[i, col])
df = df.drop(['Total'], axis=1)
df.head()
```

```python
callRate = pd.DataFrame(data={'Call_Rate': callRates, 'Nature': natures, 'Location': locs})
callRate.to_csv('call_rates.csv')
callRate.head()
```

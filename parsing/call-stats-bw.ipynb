```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import accessDB as db

pd.show_versions()
```

```python
# Read raw data in
df = db.filter(doGeoLoc=True)
```

```python
df.head()
```

```python
call_numbers = df.groupby('Call Number',as_index=False).size()
indexes_of_responded_calls = []
for key, num in call_numbers.items():
    if num > 2:
        indexes_of_responded_calls.append(key)
indexes_of_responded_calls
```

```python
# look at how the number of statuses change from ward to ward....
status_changed = df.loc[df['Call Number'].isin(indexes_of_responded_calls)]
status_changed
```

```python
districts = ['1','2','3','4','5','6','7']
status_changed = status_changed[['Call Number','Nature of Call','Status','Police District']].set_index('Call Number')
status_changed = status_changed[status_changed['Police District'].isin(districts)]
status_changed.tail(15)
```

```python
total_status_changes = status_changed.index.nunique()
total_status_changes
```

```python
total_assignment_completed = status_changed[status_changed['Status'] == 'Assignment Completed'].count()['Status']
total_assignment_completed
```

```python
total_assignment_completed/total_status_changes
```

```python
completed_assignments = status_changed[status_changed['Status'] == 'Assignment Completed']\
                            .groupby('Police District').count()['Status']
total_assignments = status_changed[~status_changed.index.duplicated(keep='first')]\
                            .groupby('Police District').count()['Status']
print(completed_assignments)
print(total_assignments)
print(completed_assignments/total_assignments)
```
